# Histogram of Oriented Gradients

What is a Feature Descriptor?

A feature descriptor is a representation of an image or an image patch that simplifies the image by extracting 
useful information and throwing away extraneous information.

# Step 1 : Preprocessing

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('/home/rjoker/DeepLearning/face_image.jpeg')
plt.imshow(img)
```

```python
img.shape
```

```python
img2 = cv2.resize(img,(152,304))
plt.imshow(img2)
```

# Step 2 : Calculate the Gradient Images

```python
# Read image
im = img2
im = np.float32(im) / 255.0

# Calculate gradient
gx = cv2.Sobel(img, cv2.CV_32F, 1, 0, ksize=1)
gy = cv2.Sobel(img, cv2.CV_32F, 0, 1, ksize=1)
```

```python
mag, angle = cv2.cartToPolar(gx, gy, angleInDegrees=True)
```

```python
plt.imshow(gx)
```

```python
plt.imshow(gy)
```
